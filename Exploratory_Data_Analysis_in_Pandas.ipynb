```python
#sort data using the total population percentage as to world total
df.sort_values(by="World Population Percentage", ascending=False)
```

```python
#correlation between numeric values / compares every column to every other column
numeric_columns = df.select_dtypes(include=['float'])
numeric_columns.corr()
```

```python
#Show correlation using heatmap
numeric_columns = df.select_dtypes(include=['float'])
sns.heatmap(numeric_columns.corr(), annot=True)
plt.rcParams['figure.figsize'] = (20, 7)
plt.show()
```
