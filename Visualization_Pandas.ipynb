# Visualizations with Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\tinot\OneDrive\Documents\My Python Repository\Pandas\Ice Cream Ratings.csv")
df
```

```python
#set date column as index column
df = df.set_index('Date')
df
```

```python
#print available themes
print(plt.style.available)
plt.style.use('tableau-colorblind10')
```

```python
#by default plot() gives us a line plot
df.plot()
```

```python
#kind = graph type, subplots = divide graph into components
df.plot(kind = 'line', subplots = True, title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
#barplot
df.plot(kind = 'bar')
```

```python
#stacked barplot
df.plot(kind = 'bar', stacked = True)
```

```python
#horizontal barchart
df.plot.barh()
```

```python
#stacked horizontal barchart
df.plot.barh(stacked = True)
```

```python
#scatter plot, s= zize of dots, c = color of dots
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s=100, c = 'pink')
```

```python
#histogram, used to show distributions of variables
df.plot.hist(bins = 10)
```

```python
#box plot
df.boxplot()
```

```python
#area plot
df.plot.area(figsize = (10,5))
```

```python
#pie chart
df.plot.pie(y='Flavor Rating',figsize=(10,10))
```
